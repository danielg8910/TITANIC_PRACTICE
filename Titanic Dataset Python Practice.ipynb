Valores faltantes:
Pueden aparecer como ? N/A 0 o simplemente una celda en blanco
Normalmente nos encontramos un NaN

Primero se debe verificar si se puede encontrar el valor faltante

Se puede eliminar datos o encontrar el valor perdido , se debe buscar hacer el menos daño posible 
Reemplazar con el promedio pero estamos haciendo una conjetura 

Pandas cuenta con el metodo dropna

axis = 0 eliminar filas
axis = 1 eliminar columnas

```python
df.dropna(subset = ["Cabina"],axis = 0) # The parameter subset filter the name of colum for erase de NaN values
```

```python
df.dropna(axis = 1) # This option erase de fill with Nan values , the parameter inplace = True does the changes in the dataframe directly
```

```python
promedio = round(df["Edad"].mean(),0)
```

```python
df["Edad"].replace(np.nan,promedio) #Replace method changes the values with NaN by the choice values
```

```python
df.tail()
```

The pd.get_dummies() method convert the categoric variables in numeric variables 


```python
pd.get_dummies(df, columns = ["Sexo"],drop_first ="True") #Drop_first eliminate the second column and the variable is 1 for men 0 for woman
```

```python
bins = [0,5,12,18,35,60,100] #in this moment will be assignment the bins for the cut module, that are the ages range
```

```python
names = ["1","2","3","4","5","6"] #the name for the groups
```

```python
df["Edad"] =pd.cut(df["Edad"],bins,labels = names)  #in the cut module define bins and labels
```
